## Câu hỏi 6: Liệt kê các trận đấu và thông tin về leage và country tương ứng

Bảng Match có quan hệ với :
 - Bảng Country thông qua khóa country_id
 - Bảng League thông qua khóa League_id
 
Sử dụng câu lệnh join để kết nối dư liệu từ 3 bảng trên.


```python
# ĐIỀN CODE VÀO ĐÂY
result = pd.read_sql("SELECT M.*, L.name AS league_name, C.name AS country_name " \
                     "FROM Match M " \
                     "JOIN League L ON M.league_id = L.id " \
                     "JOIN Country C ON L.country_id = C.id", conn)
result
```

## Câu hỏi 7: Liệt kê số lượng trận đấu của mỗi giải đấu bao gồm cả tên giải đấu, sắp xếp theo thứ tự giảm dần số trận đấu

```python
# ĐIỀN CODE VÀO ĐÂY
result = pd.read_sql("SELECT L.name AS League_name, COUNT(M.id) AS Total_matches " \
                     "FROM Match M " \
                     "JOIN League L ON M.league_id = L.id " \
                     "GROUP BY L.id " \
                     "ORDER BY total_matches DESC", conn)
result
```

## Câu hỏi 8: Liệt kê tổng số bàn thắng của đội nhà và đội khách trong mỗi giải đấu

## Câu hỏi 10: Liệt kê 20 đội với số bàn thắng sân nhà cao nhất
- Sử dung câu lênh GROUP BY, LIMIT

```python
# ĐIỀN CODE VÀO ĐÂY
result = pd.read_sql("SELECT T.team_long_name, " \
                        "SUM(M.home_team_goal) AS home_goals " \
                    "FROM Team T " \
                    "JOIN Match M ON T.team_api_id = M.home_team_api_id " \
                    "GROUP BY T.team_api_id " \
                    "ORDER BY home_goals DESC LIMIT 20", conn)
result
```

## Câu hỏi 11: Liệt kê 20 đội có số bàn thắng trên sân khách cao nhất
- Sử dụng câu lênh GROUP BY, LIMIT

```python
# ĐIỀN CODE VÀO ĐÂY
result = pd.read_sql("SELECT T.team_long_name, " \
                        "SUM(M.away_team_goal) AS away_goals " \
                    "FROM Team T " \
                    "JOIN Match M ON T.team_api_id = M.away_team_api_id " \
                    "GROUP BY T.team_api_id " \
                    "ORDER BY away_goals DESC LIMIT 20", conn)
result
```

## Câu hỏi 12: Liệt kê tên các đội bóng và tổng số bàn thắng mỗi đội, sắp xếp  theo số lượng giảm dần

- Tổng số bàn thắng của một đội bao gồm cả số bàn thắng ở sân nhà và sân khách.
- Sử dụng câu lệnh Union  

```python
# ĐIỀN CODE VÀO ĐÂY
result = pd.read_sql("SELECT T.team_long_name, " \
                        "SUM(goals) AS total_goals " \
                    "FROM ( SELECT M.home_team_api_id AS team_id, M.home_team_goal AS goals " \
                            "FROM Match M " \
                            "UNION " \
                            "SELECT M.away_team_api_id AS team_id, M.away_team_goal AS goals " \
                            "FROM Match M ) " \
                    "AS G " \
                    "JOIN Team T ON T.team_api_id = G.team_id " \
                    "GROUP BY T.team_api_id " \
                    "ORDER BY total_goals DESC", conn)
result
```

## Câu hỏi 13: Liệt kê tên các đội bóng (long name) và tổng số trận đấu đội đó tham gia, sắp xếp theo thứ tự giảm dần.

```python
# ĐIỀN CODE VÀO ĐÂY
result = pd.read_sql("SELECT T.team_long_name, " \
                        "COUNT(M.id) AS matches_played " \
                     "FROM Team T " \
                     "LEFT JOIN Match M  ON T.team_api_id = M.home_team_api_id  OR T.team_api_id = M.away_team_api_id " \
                     "GROUP BY T.team_api_id " \
                     "ORDER BY matches_played DESC", conn)
result
```

## Câu hỏi 14: Liệt kê số trận thắng, thua và hòa của mỗi đội.

- Nếu một đội có số lượng bàn thắng trên sân nhà > số lượng bàn thắng trên sân khách của chính đội đó, thì đội đó sẽ "thắng" trong trận đấu này
- Nếu một đội có số lượng bàn thắng trên sân nhà < số lượng bàn thắng trên sân khách của chính đội đó, thì đội đó sẽ "thua" trong trận đấu này
- Nếu một đội có số lượng bàn thắng trên sân nhà = số lượng bàn thắng trên sân khách của chính đội đó, thì đội đó sẽ "hòa" trong trận đấu này
- Sử dụng câu lệnh "Case When"


```python
# ĐIỀN CODE VÀO ĐÂY
query = """
SELECT 
    T.team_long_name,
    SUM(
      CASE
        WHEN M.home_team_api_id = T.team_api_id AND M.home_team_goal  > M.away_team_goal THEN 1
        WHEN M.away_team_api_id = T.team_api_id AND M.away_team_goal  > M.home_team_goal THEN 1
        ELSE 0
      END
    ) AS wins,
    SUM(
      CASE
        WHEN M.home_team_api_id = T.team_api_id AND M.home_team_goal  < M.away_team_goal THEN 1
        WHEN M.away_team_api_id = T.team_api_id AND M.away_team_goal  < M.home_team_goal THEN 1
        ELSE 0
      END
    ) AS losses,
    SUM(
      CASE
        WHEN (M.home_team_api_id = T.team_api_id OR M.away_team_api_id = T.team_api_id)
             AND M.home_team_goal = M.away_team_goal
        THEN 1 ELSE 0
      END
    ) AS draws
FROM Team T
LEFT JOIN "Match" M
  ON T.team_api_id = M.home_team_api_id
  OR T.team_api_id = M.away_team_api_id
GROUP BY T.team_api_id
ORDER BY wins DESC;
"""

# chạy và xem kết quả
result = pd.read_sql(query, conn)
result
```

## Câu hỏi 15: Liệt kê 10 đội có số trận thắng nhiều nhất

```python
# ĐIỀN CODE VÀO ĐÂY
query15 = """
SELECT * FROM (
    SELECT 
        T.team_long_name,
        SUM(CASE 
            WHEN M.home_team_api_id = T.team_api_id AND M.home_team_goal > M.away_team_goal THEN 1
            WHEN M.away_team_api_id = T.team_api_id AND M.away_team_goal > M.home_team_goal THEN 1
            ELSE 0 END) AS wins
    FROM Team T
    LEFT JOIN Match M 
      ON T.team_api_id = M.home_team_api_id OR T.team_api_id = M.away_team_api_id
    GROUP BY T.team_api_id
) ORDER BY wins DESC
LIMIT 10;
"""

# chạy và xem kết quả
result = pd.read_sql(query15, conn)
result
```

## Câu hỏi 16: Liệt kê tỉ lệ phần trăm về số trận của mỗi giải trên tổng tất cả các trận

```python
# ĐIỀN CODE VÀO ĐÂY
result = pd.read_sql("SELECT L.name, " \
                        "COUNT(M.id) * 100.0 / (SELECT COUNT(*) FROM Match) AS percent_matches " \
                     "FROM Match M " \
                     "JOIN League L ON M.league_id = L.id " \
                     "GROUP BY L.id;", conn)
result
```

## Câu hỏi 17: Liệt kê tỉ lệ phần trăm về số bàn ghi được của từng giải so với tổng số bàn của tất cả các trận.

```python
# ĐIỀN CODE VÀO ĐÂY
result = pd.read_sql("SELECT L.name, (SUM(M.home_team_goal + M.away_team_goal)) * 100.0 /  (SELECT SUM(home_team_goal + away_team_goal) FROM Match) AS percent_goals " \
                     "FROM Match M " \
                     "JOIN League L ON M.league_id = L.id " \
                     "GROUP BY L.id;", conn)
result
```

## Câu hỏi 18: Liệt kê tổng số bàn ghi được của từng giải đấu theo từng mùa giải.

```python
# ĐIỀN CODE VÀO ĐÂY
result = pd.read_sql("SELECT L.name AS league_name, M.season, " \
                        "SUM(M.home_team_goal + M.away_team_goal) AS total_goals " \
                     "FROM Match M JOIN League L ON M.league_id = L.id GROUP BY L.id, M.season " \
                     "ORDER BY M.season, league_name;", conn)
result
```

## Câu hỏi 19: Liệt kê thông tin cầu thủ:

- Quy đổi cân nặng sang kilogram

- Tính chỉ số bmi = ( (weight* 0.453592) / (height/100)^2)
- Tính tuổi của cầu thủ

```python
# ĐIỀN CODE VÀO ĐÂY
result = pd.read_sql("SELECT P.player_name, " \
                     "P.birthday, "
                     "(P.weight * 0.453592) AS weight_kg, "
                     "(P.height / 100.0) AS height_m, "
                     "(P.weight * 0.453592) / ((P.height / 100.0) * (P.height / 100.0)) AS BMI, "
                     "(CAST(strftime('%Y', 'now') AS INT) - CAST(strftime('%Y', P.birthday) AS INT)) AS age " \
                     "FROM Player P;", conn)
result
```

## Câu hỏi 21: Liệt kê  cầu thủ có số trận đấu nhiều nhất

```python
# ĐIỀN CODE VÀO ĐÂY
query21 = """
SELECT P.player_name,
       COUNT(M.id) AS matches_played
FROM Player P
JOIN Player_Attributes PA ON P.player_api_id = PA.player_api_id
JOIN Match M 
   ON P.player_api_id IN (
        M.home_player_1, M.home_player_2, M.home_player_3, M.home_player_4, M.home_player_5,
        M.home_player_6, M.home_player_7, M.home_player_8, M.home_player_9, M.home_player_10, M.home_player_11,
        M.away_player_1, M.away_player_2, M.away_player_3, M.away_player_4, M.away_player_5,
        M.away_player_6, M.away_player_7, M.away_player_8, M.away_player_9, M.away_player_10, M.away_player_11
    )
GROUP BY P.player_api_id
ORDER BY matches_played DESC
LIMIT 1;
"""

# chạy và xem kết quả
result = pd.read_sql(query21, conn)
result
```

## Câu hỏi 22: Liệt kê các cầu thủ có tổng số điểm đánh giá lớn hơn 80

```python
# ĐIỀN CODE VÀO ĐÂY
result = pd.read_sql("SELECT P.player_name, " \
                        "SUM(PA.overall_rating) AS total_rating " \
                     "FROM Player P " \
                     "JOIN Player_Attributes PA ON P.player_api_id = PA.player_api_id " \
                     "GROUP BY P.player_api_id HAVING total_rating > 80;", conn)
result
```
